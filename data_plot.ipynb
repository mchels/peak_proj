```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bg_vals = np.loadtxt('backgate_values.txt')
mag_vals = np.loadtxt('mag_values.txt')
cond_vals = np.loadtxt('cond_values.txt')

bg_label = 'Backgate voltage (V_BG)'
mag_label = 'Magnetic field magnitude (Tesla)'
cond_label = 'Differential conductance dI/dV (e^2/h)'
```

# Plot the full data set

```python
fig, ax = plt.subplots(figsize=(10,7))
im = ax.pcolormesh(mag_vals, bg_vals, cond_vals)
mag_field = -0.5 # Magnetic field value for next plot.
ax.axvline(mag_field, color='C1')
ax.set_xlabel(mag_label)
ax.set_ylabel(bg_label)
cbar = plt.colorbar(mappable=im)
cbar.set_label(cond_label)
plt.show()
```

# Make 1D plot at orange line above

```python
idx = np.searchsorted(mag_vals, mag_field)
cond_1D = cond_vals[:,idx]
fig, ax = plt.subplots()
ax.plot(bg_vals, cond_1D)
ax.set_xlabel(bg_label)
ax.set_ylabel(cond_label)
plt.show()
```
